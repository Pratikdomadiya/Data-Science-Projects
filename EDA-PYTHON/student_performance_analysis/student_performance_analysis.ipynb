# cleaning the data

```python
#check first for null values and fix it.
print(data.isnull().sum())
print(data.isna().sum())
# print(data.isin("none").any())
```

```python

#remove unnecessary columns 
student= data.drop(['race/ethnicity'],axis=1)
```

```python
student.head()
```

```python
#relationship analysis
correlation=student.corr()
sns.heatmap(correlation , xticklabels=correlation.columns,yticklabels=correlation.columns,
           annot=True,)
```

```python
#show relationship between each and every variable 
sns.pairplot(student)
```

```python
#scatter plot is used for two numerical variables relationships
sns.relplot(x="math score",y="reading score",hue='gender',data=student)
```

```python
sns.relplot(x="math score",y="reading score",hue='lunch',data=student)
```

```python
#histogram- display range of observation
sns.distplot(student['math score'])
```

```python
sns.distplot(student['reading score'])
```

```python
sns.distplot(student['writing score'],bins=5)
```

```python
# fig, axs = plt.subplots(nrows=3)
sns.catplot(x='math score',kind='box',data=student)
#here most of the values are present between around 58 to 77.
# sns.catplot(x='reading score',kind='box',data=student,axs=axs[1])
# sns.catplot(x='writing score',kind='box',data=student,axs=axs[2])
```

```python
fig, axs = plt.subplots(nrows=3,figsize=(15,15))
sns.barplot(x='parental level of education',y='reading score',data=student,ax=axs[0])
sns.barplot(x='parental level of education',y='math score',data=student,ax=axs[1])
sns.barplot(x='parental level of education',y='writing score',data=student,ax=axs[2])
```

```python
fig, axs = plt.subplots(nrows=3,figsize=(15,15))
sns.barplot(x='gender',y='reading score',data=student,ax=axs[0])
sns.barplot(x='gender',y='math score',data=student,ax=axs[1])
sns.barplot(x='gender',y='writing score',data=student,ax=axs[2])
```

```python
fig, axs = plt.subplots(nrows=3,figsize=(15,15))
sns.barplot(x='lunch',y='reading score',data=student,ax=axs[0])
sns.barplot(x='lunch',y='math score',data=student,ax=axs[1])
sns.barplot(x='lunch',y='writing score',data=student,ax=axs[2])
# sns.barplot(x='lunch',y=(student[student['gender']=='male']).value_counts(),data=student,ax=axs[3])
```
